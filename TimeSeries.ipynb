```python
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Date": pd.to_datetime(['2022-01-01','2022-02-01','2022-03-01','2022-04-01','2022-05-01','2022-06-01']),
    "Value": (10,13,20,48,67,78)
}
df=pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"],marker='X')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with Trends")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('/content/AirPassengers.csv')
data['Month'] = pd.to_datetime(data['Month'], format='%Y-%m')
data.set_index('Month', inplace=True)

# Plot the original time series data
plt.figure(figsize=(10, 4))
plt.plot(data, label='Original Time Series')
plt.title('Time series with seasonality')
plt.xlabel('Year')
plt.ylabel('Number of Passengers')
plt.legend()
plt.show()
```
