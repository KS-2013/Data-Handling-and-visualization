MATRIX VISUALIZATIION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=np.random.randn(13,16)
plt.imshow(data,cmap='magma')
plt.title("2D Heat map")
plt.colorbar()
plt.show()
```

SEABORN

```python
import seaborn as sns
data=np.random.randint(low=14,
                       high=100,
                       size=(10,10))
hm=sns.heatmap(data=data,annot=True)
plt.show()
```

TREE VISUALIZATION (TREE MAPS)

```python
import plotly.express as px
df=px.data.tips()
fig=px.treemap(df,path=[px.Constant("all"),'sex','day','time'],
               values='total_bill',color='day')
fig.update_layout(margin=dict(t=50,l=25,r=25,b=25))
fig.show()
```
