```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter
```

```python
text_data="Social media is a reality of day-to-day life – for many, it’s the lens through which lives are conducted and viewed. So it’s quite the shift in perspective to see social media instead as the world’s largest pool of text data"
wordcloud= WordCloud(width=800,height=400,background_color="black").generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.title("WordCloud")
plt.show()
words=text_data.split()
word_Counter=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_Counter.keys(), word_Counter.values())
plt.title("Histogram of word frequencies")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=90)
plt.show()
```
